```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data/sudan.csv")
df.head()
```

```python
df = df.replace(-999, np.nan)
```

```python
df["Country"] = "Sudan"
df.head()
```

```python
df["DATE"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
```

```python
df["Month"] = df["DATE"].dt.month
```

```python
df.isna().sum()
```

```python
from scipy.stats import zscore

cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M"]

z = df[cols].apply(zscore)
outliers = (abs(z) > 3)

outliers.sum()
```

```python
df.groupby("Month")["T2M"].mean().plot()
plt.title("Monthly Temperature Trend")
plt.show()
```

```python
df.groupby("Month")["PRECTOTCORR"].sum().plot(kind="bar")
plt.title("Monthly Rainfall")
plt.show()
```

```python
sns.heatmap(df[cols].corr(), annot=True)
plt.show()
```

```python
plt.scatter(df["T2M"], df["RH2M"])
plt.title("Temperature vs Humidity")
plt.show()
```

```python
df.to_csv("data/ethiopia_clean.csv", index=False)
```
