```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data/sudan.csv")
df.head()
```

```python
df.describe()
```

The dataset contains 4,108 observations from 2015 to 2026, representing a long-term view of climate conditions.The average temperature (T2M) is approximately 28.76°C, indicating a hot climate overall. Maximum temperatures reach up to 45.96°C, showing periods of extreme heat, while minimum temperatures drop to around 5.93°C, indicating strong variability across seasons or regions.The temperature range (T2M_RANGE) averages about 15.27°C, suggesting a large difference between daytime and nighttime temperatures.Precipitation is generally very low, with a mean of 0.64 and a median of 0, indicating that most days are dry. However, the maximum value (66.49) shows that occasional heavy rainfall events occur. This distribution suggests a highly skewed rainfall pattern, where dry conditions dominate but rare intense rainfall events significantly affect the dataset.Wind speed averages 3.48 m/s, with maximum values reaching 7.15 m/s, indicating moderate wind activity. The variability suggests occasional stronger wind events but generally stable wind conditions.

```python
df = df.replace(-999, np.nan)
```

```python
df["Country"] = "Sudan"
df.head()
```

```python
df["DATE"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df.head()
```

```python
df["Month"] = df["DATE"].dt.month
df.head()
```

```python
df.isna().sum()
```

no missing values

```python
from scipy.stats import zscore

cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M"]

z = df[cols].apply(zscore)
outliers = (abs(z) > 3)

outliers.sum()
outlier_count = outliers.any(axis=1).sum()
print("Number of rows with |Z| > 3:", outlier_count)
```

```python
df.isnull().sum()
```

no missing values

```python
df.groupby("Month")["T2M"].mean().plot()
plt.title("Monthly Temperature Trend")
plt.show()
```

5th month is the hottest month
2nd month is the coolest month.

```python
df.groupby("Month")["PRECTOTCORR"].sum().plot(kind="bar")
plt.title("Monthly Rainfall")
plt.show()
```

The 8th month is the pick rainy season.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["T2M_RANGE"], df["WS2M"], alpha=0.5)
plt.title("T2M_RANGE vs WS2M")
plt.xlabel("Temperature Range")
plt.ylabel("Wind Speed")
plt.show()
corr_unstacked = corr.abs().unstack()
corr_unstacked = corr_unstacked[corr_unstacked < 1]  # remove self-correlation
top_3 = corr_unstacked.sort_values(ascending=False).head(3)
print(top_3)
```

DOY and Month (r ≈ 0.9966)
There is an almost perfect positive correlation between Day of Year (DOY) and Month.They carry almost identical information about the calendar cycle.
T2M and T2M_MIN (r ≈ 0.9419)
There is a very strong positive correlation between average temperature (T2M) and minimum temperature (T2M_MIN).

```python
plt.scatter(df["T2M"], df["RH2M"])
plt.title("Temperature vs Humidity")
plt.show()
```
